```python
#importar as bibliotecas necessárias 
from selenium import webdriver as opcoesSelenium
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pyautogui as tempoEspera
from openpyxl import load_workbook
import os

#caminho do arquivo em excel no computador
nomeCaminhoArquivo = 'C:\\Users\ADMIN\Documents\Python RPA e Excel como automatizar Processos e Planilhas!\SEÇAO 5 - ROBÔ 4 - PREENCHENDO FORMULARIOS WEB COM PYTHON\DadosFormulario.xlsx'
planilhaAberta = load_workbook(filename = nomeCaminhoArquivo)

#selecionar sheet que tem as informações a serem passadas
sheetSelecionado = planilhaAberta['Dados']

for linha in range(2, len(sheetSelecionado['A']) + 1):
    
    nome = sheetSelecionado['A%s' % linha].value
    email = sheetSelecionado['B%s' % linha].value
    telefone = sheetSelecionado['C%s' % linha].value
    sexo = sheetSelecionado['D%s' % linha].value
    sobre = sheetSelecionado['E%s' % linha].value

    #tempo Espera
    tempoEspera.sleep(2)

    navegadorFormulario = opcoesSelenium.Chrome()
    navegadorFormulario.get('https://pt.surveymonkey.com/r/Y9Y6FFR')

    tempoEspera.sleep(2)

    #preenche nome
    navegadorFormulario.find_element(By.NAME,'683928983').send_keys(nome)

    tempoEspera.sleep(1)

    #preenche e-mail
    navegadorFormulario.find_element(By.NAME,'683932318').send_keys(email)

    tempoEspera.sleep(1)

    #preenche telefone
    navegadorFormulario.find_element(By.NAME,'683930688').send_keys(telefone)

    tempoEspera.sleep(1)

    #preenche escreva algo sobre você
    navegadorFormulario.find_element(By.NAME,'683932969').send_keys(sobre)

    tempoEspera.sleep(1)
    
    if sexo == 'Masculino':
        #preenche radio button masculino 
        navegadorFormulario.find_element(By.ID,'683931881_4497366118_label').click()
    else:
        #preenche radio button feminino 
        navegadorFormulario.find_element(By.ID,'683931881_4497366119_label').click()

    tempoEspera.sleep(2)

    #clica para enviar informações
    navegadorFormulario.find_element(By.XPATH, '//*[@id="patas"]/main/article/section/form/div[2]/button').click()





```
